```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style('whitegrid')
%matplotlib inline
```

**Explanation:**  
We import necessary libraries:  
- `pandas` for data manipulation  
- `matplotlib` and `seaborn` for visualization  
- `numpy` for numerical operations  
We also set seaborn style for cleaner plots.

```python
data_path = "../data/Customer-Churn.csv"
df = pd.read_csv(data_path)
df.head()
```

**Explanation:**  
Loaded the raw dataset from the `data` folder and displayed the first 5 rows to inspect the columns and sample data.


```python
print("Dataset Shape:", df.shape)
df.info()
df.describe(include='all')
```

**Explanation:**  
We checked:  
- Dataset shape (rows x columns)  
- Column data types and null values  
- Summary statistics for numeric and categorical columns to understand data distribution


```python
missing = df.isnull().sum()
print("Missing Values per Column:\n", missing)
```

**Explanation:**  
We checked for missing values using `isnull()` and visualized them with a heatmap.  
This helps identify incomplete columns or rows that may need cleaning before modeling.

```python
sns.countplot(data=df, x='Churn')
plt.title("Churn Distribution")
plt.show()

churn_rate = df['Churn'].value_counts(normalize=True) * 100
print("Churn Percentage:\n", churn_rate)
```

**Explanation:**  
Analyzed the target variable `Churn`:  
- Countplot shows the number of churned vs non-churned customers  
- Calculated percentage to see class imbalance (important for modeling strategy)


```python
numeric_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.show()
```

**Explanation:**  
Visualized numeric features (`tenure`, `MonthlyCharges`, `TotalCharges`) to:  
- Understand distributions  
- Detect outliers or skewness  
- Decide on scaling or transformation in the next step


```python
categorical_cols = ['gender', 'Partner', 'Dependents', 'PhoneService', 
                    'MultipleLines', 'InternetService', 'OnlineSecurity', 
                    'OnlineBackup', 'DeviceProtection', 'TechSupport', 
                    'StreamingTV', 'StreamingMovies', 'Contract', 
                    'PaperlessBilling', 'PaymentMethod']

for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x=col, hue='Churn')
    plt.title(f'{col} vs Churn')
    plt.xticks(rotation=45)
    plt.show()
```

**Explanation:**  
Visualized categorical features against `Churn` to identify patterns:  
- Some categories may have higher churn rates  
- Helps in feature selection and engineering


```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.dropna(inplace=True)
df['Churn'] = df['Churn'].map({'Yes':1, 'No':0})
df.to_csv("../data/cleaned_data.csv", index=False)
print("Cleaned data saved to ../data/cleaned_data.csv")
```

**Explanation:**  
- Converted `TotalCharges` to numeric and removed missing values  
- Converted `Churn` to 0/1 for modeling  
- Saved the cleaned dataset for the next notebook: Feature Engineering
